## ML Model
It's a very basic ML model which works on the  princiiple that with successive normalisation of a matrix it converges to the eigen vector with the largest eigen value

We will start with different people in different areas in college "Hostels", "Mess", "Department" and we will realize that after a certain time it converges to a certain value, and that values is same no matter where you start from.
here we start with everyone in Hostel.

```python
H=3000
M=0
D=0
```

```python
import random
import matplotlib.pyplot as plt
Hostel=[]
Mess=[]
Department=[]
Hostel.append(H)
Mess.append(M)
Department.append(D)
```

Now we assign probabilities, 
For e.g "HpD" is the probability of a person going to Department from Hostel

```python
HpD=0.7
DpM=0.5
MpD=0.4
tH=0
tM=0
tD=0
```

Now, Let's run this and simulate it for 1000 Days and see where students are

```python
for i in range(1000):
    for i in range(H):
        p=random.uniform(0,1)
        if(p<HpD):
            tD+=1
        else:
            tM+=1
    for i in range(M):
        p=random.uniform(0,1)
        if(p<MpD):
            tD+=1
        else:
            tH+=1
    for i in range(D):
        p=random.uniform(0,1)
        if(p<DpM):
            tM+=1
        else:
            tH+=1
    H=tH
    M=tM
    D=tD
    tH=0
    tM=0
    tD=0
    Hostel.append(H)
    Mess.append(M)
    Department.append(D)
```

```python
day=list(range(1001))
plt.xlabel("Day")
plt.ylabel("Number of People")
plt.ylim(0,3000)
plt.xlim(0,50)

plt.plot(Hostel)
plt.plot(Mess)
plt.plot(Department)
plt.show()
```

Here we can see that the value converges after few days and now let's perform the same thing with different initital values

```python
H=1000
M=1000
D=1000
Hostel1=[]
Mess1=[]
Department1=[]
Hostel1.append(H)
Mess1.append(M)
Department1.append(D)
for i in range(1000):
    for i in range(H):
        p=random.uniform(0,1)
        if(p<HpD):
            tD+=1
        else:
            tM+=1
    for i in range(M):
        p=random.uniform(0,1)
        if(p<MpD):
            tD+=1
        else:
            tH+=1
    for i in range(D):
        p=random.uniform(0,1)
        if(p<DpM):
            tM+=1
        else:
            tH+=1
    H=tH
    M=tM
    D=tD
    tH=0
    tM=0
    tD=0
    Hostel1.append(H)
    Mess1.append(M)
    Department1.append(D)
day=list(range(1001))
plt.xlabel("Day")
plt.ylabel("Number of People")
plt.ylim(0,3000)
plt.xlim(0,50)

plt.plot(Hostel1)
plt.plot(Mess1)
plt.plot(Department1)
plt.show()
```

Here we can see that it doesn't matter what initial values will take, for a given probability distribution it will converge at the same value.
