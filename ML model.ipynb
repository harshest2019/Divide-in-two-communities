```python
import random
import matplotlib.pyplot as plt
Hostel=[]
Mess=[]
Department=[]
H=3000
M=0
D=0
Hostel.append(H)
Mess.append(M)
Department.append(D)
HpD=0.7
DpM=0.5
MpD=0.4
tH=0
tM=0
tD=0

for i in range(1000):
    for i in range(H):
        p=random.uniform(0,1)
        if(p<HpD):
            tD+=1
        else:
            tM+=1
    for i in range(M):
        p=random.uniform(0,1)
        if(p<MpD):
            tD+=1
        else:
            tH+=1
    for i in range(D):
        p=random.uniform(0,1)
        if(p<DpM):
            tM+=1
        else:
            tH+=1
    H=tH
    M=tM
    D=tD
    tH=0
    tM=0
    tD=0
    Hostel.append(H)
    Mess.append(M)
    Department.append(D)
```

```python
day=list(range(1001))
plt.xlabel("Day")
plt.ylabel("Number of People")
plt.ylim(0,3000)
plt.xlim(0,100)

plt.plot(Hostel)
plt.plot(Mess)
plt.plot(Department)
plt.show()
```
